### Gr√°ficos de linha em paralelo

```python
x = linspace(0, 5, 10)
y = x ** 2

fig, axes = plt.subplots(nrows = 1, ncols = 2) #divide a √°rea de plotagem em dois subplots

for ax in axes: #loop pelos eixos para criar cada plot
    ax.plot(x, y, 'r')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('T√≠tulo')

fig.tight_layout()

plt.show()
```

### Gr√°ficos de linha com diferentes escalas

```python
x = linspace(0, 5, 10)
y = x ** 2

fig, axes = plt.subplots(1, 2, figsize = (10, 4))

axes[0].plot(x, x ** 2, x, exp(x))
axes[0].set_title("Escala Padr√£o")

axes[1].plot(x, x**2, x, exp(x))
axes[1].set_yscale("log")
axes[1].set_title("Escala Logar√≠tmica (y)")

plt.show()
```

### Gr√°fico com Grid

```python
x = linspace(0, 5, 10)
y = x ** 2

fig, axes = plt.subplots(1, 2, figsize = (10, 3))

axes[0].plot(x, x ** 2, x ** 3, lw = 2)
axes[0].grid(True)

axes[1].plot(x, x ** 2, x, x ** 3, lw = 2)
axes[1].grid(color = 'b', alpha = 0.7, linestyle = 'dashed', linewidth = 0.8)

plt.show()
```

### Diferentes estilos de Plots

```python
xx = np.linspace(-0.75, 1., 100)
n = np.array([0, 1, 2, 3, 4, 5])

fig, axes = plt.subplots(1, 4, figsize = (12, 3)) #uma linha com 4 colunas

axes[0].scatter(xx, xx + 0.25 * randn(len(xx)), color = "black")
axes[0].set_title("scatter")

axes[1].step(n, n ** 2, lw = 2, color = "blue")
axes[1].set_title("Step")

axes[2].bar(n, n ** 2, align = "center", width = 0.5, alpha = 0.5, color = "magenta")
axes[2].set_title("Bar")

axes[3].fill_between(x, x ** 2, x ** 3, alpha = 0.5, color = "green")
axes[3].set_title("Fill Between")


plt.show()
```

### Histogramas
Utilizado para visualizar a distribui√ß√£o de uma vari√°vel cont√≠nua. Eles s√£o compostos por barras retangulares adjacentes, onde a √°rea de cada barra √© proporcional √† frequ√™ncia de observa√ß√µes de dados que caem em uma faixa espec√≠fica de valores.

O eixo horizontal representa os intervalos de classe, enquanto o eixo vertical representa a frequ√™ncia de observa√ß√µes de dados que caem em cada intervalo de classe. Os intervalos de classe devem ser escolhidos de forma apropriada para a distribui√ß√£o dos dados e a escala dos eixos deve ser definida adequadamente.

```python
n = np.random.randn(100000) #randn = conjunto de dados com valores aleat√≥rios

fig, axes = plt.subplots(1, 2, figsize = (12, 4))

axes[0].hist(n)
axes[0].set_title("Histograma Padr√£o")
axes[0].set_xlim((min(n), max(n))) #defino os limites

axes[1].hist(n, cumulative = True, bins = 50)
axes[1].set_title("Histograma Cumulativo")
axes[1].set_xlim((min(n), max(n)))


plt.show()
```

### Gr√°ficos 3D

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D
```

```python
alpha = 0.7
phi_ext = 2 * np.pi * 0.5

def ColorMap(phi_m, phi_p):
    return( + alpha - 2 * np.cos(phi_p)*cos(phi_m) - alpha * np.cos(phi_ext - 2*phi_p))

phi_m = np.linspace(0, 2*np.pi, 100)
phi_p = np.linspace(0, 2*np.pi, 100)
X, Y = np.meshgrid(phi_p, phi_m)
Z = ColorMap(X, Y).T
```

```python
fig = plt.figure(figsize = (14, 6))

#Adiciona o  subplot 1 com proje√ß√£o 3D
ax = fig.add_subplot(1, 2, 1, projection = '3d')
p = ax.plot_surface(X, Y, Z, rstride = 4, cstride = 4, linewidth = 0)

#Adiciona o  subplot 2 com proje√ß√£o 3D
ax = fig.add_subplot(1, 2, 2, projection = '3d')
p = ax.plot_surface(X, Y, Z, rstride = 1, cstride = 1, cmap = cm.coolwarm, linewidth = 0, antialiased = False)

#Cria a barra de cores como legenda
cb = fig.colorbar(p, shrink = 0.5)


plt.show()
```
