### Agrupamento de Dados em Data Frames com Group By

A função Group By é uma funçãoque ajuda a obter uma visão geral dos dados, tornando mais fácil a exploração do conjunto de dados e revelar os relacionamentos entre as variáveis.

```python
df[['Segmento', 'Regiao', 'Valor_Venda']].groupby(['Segmento', 'Regiao']).mean()
```

Primeiro filtramos os dados extraindo 3 colunas = Segmento, Regiao e Valor_Venda.
Na sequencia agrupamentos por duas colunas = Segmento e Regiao
E entao calculamos a média da coluna que não aparece no Group By

### Agregação Múltipla com a função agg()

```python
df[['Segmento', 'Regiao', 'Valor_Venda']].groupby(['Segmento', 'Regiao']).agg(['mean', 'std', 'count'])
```

Agora agregamos os dados calculando a média, o desvio padrão, e a contagem de elementos para a coluna que ficou fora do group by. A função agg() recebe como argumento uma lista de funções para agregação.

### Split de Strings em DataFrame

```python
df['ID_Pedido'].head()
```

Temos o país, o ano e o id do pedido. Vamos dividir essa coluna

```python
df['ID_Pedido'].str.split('-') #dividindo a coluna pelo - , entregando uma lista de valores
```

```python
df['ID_Pedido'].str.split('-').str[1].head()
```

Não aparece no resultado anterior, mas assim como a indexação das linhas em Python começa com 0, as colunas também. Dessa forma: País é a coluna 0, ano a coluna 1 e id coluna 2. Como queremos o ano, no [] de str, colocamos 1

```python
#Agora criamos uma nova coluna no df
df['Ano'] = df['ID_Pedido'].str.split('-').str[1]
```

```python
df.head()
```

### Strip de Strings em DataFrame

A função strip() REMOVE caracteres da string

```python
df['Data_Pedido'].head(3)
```

Temos a data de envio do produto no formato YYYY-MM-DD. E se quisermos deixar o ano com apenas 2 números, mas sem alterar o tipo de variável?

```python
df['Data_Pedido'].str.lstrip('20') #lstrip ou left strip, remove os caracteres a partir do lado esquerdo da string
```

Como não usamos o inplace = True a mudança é somente na memória e não altera o dataframe

### Replace de Strings em DataFrame

E se ao invés de remover, nós quisermos SUBSTITUIR caracteres dentro de uma string?

```python
df['ID_Cliente'] = df['ID_Cliente'].str.replace('CG', 'AX')
```

```python
df.head()
```

Substituimos o código CG por AX, na coluna ID_Cliente

### Combinação de Strings

A função cat() pode ser usada para concatenar strings

```python
df['Pedido_Segmento'] = df['ID_Pedido'].str.cat(df['Segmento'], sep = '-')
```

Criando uma nova variável, combinando dados de outras variáveis. Nesse caso, combinamos o ID Pedido e o Segmento, e entre esses valores colocamos o -
