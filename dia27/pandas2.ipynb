### Query de Dados

```python
import pandas as pd
```

```python
#Checando os valores mínimo e máximo da coluna Valor_Venda
df = pd.read_csv("dataset.csv")
df.head()
```

```python
df.Valor_Venda.describe()
```

```python
# Agora gerando uma consulta a partir de um intervalo entre 230 e 11000
df1 = df.query('230 < Valor_Venda < 11000')
```

```python
df1.Valor_Venda.describe()
```

```python
# Gerando um novo dataframe apenas com os valores de venda acima da média
df2 = df1.query('Valor_Venda > 771')
df2.head()
```

### Verificando a Ocorrência de Diversos Valores em uma coluna
O Pandas utiliza o método isin() para checar diversos valores em um coluna. É o mesmo método que a cláusula IN no SQL

```python
df.shape
```

```python
df[df['Quantidade'].isin([5, 7, 9, 11])] #Imagine que precisamos saber em quais vendas foram vendidosw 5, 7, 9 ou 11 itens
```

```python
df[df['Quantidade'].isin([5, 7, 9, 11])].shape
```

```python
df[df['Quantidade'].isin([5, 7, 9, 11])][:10] #para visualizar somente os 10 primeiros
```

### Operadores Lógicos

Utilizando o operador lógico AND (&) para checar duas condições. Serão retornados os registros quando as duas condições forem verdadeiras

```python
df[(df.Segmento == 'Home Office') & (df.Regiao == 'South')].head()
```

Mas se for necessário checar duas condições e retornar os registros se PELO MENOS UMA FOR VERDADEIRA. Nesse caso usamos o operador OR ( | )

```python
df[(df.Segmento == 'Home Office') | (df.Regiao == 'South')].tail() #parte final do df
```

Para verificar quando nenhuma condição é verdadeira !=

```python
df[(df.Segmento != 'Home Office') & (df.Regiao != 'South')].sample(5) #sample mostra os resultados de forma aleatoria
```
