### Usando Matplotlib, Seaborn, Numpy e Pandas na Cria√ß√£o de Gr√°fico Estat√≠stico

```python
np.random.seed(42)
n = 1000
pct_smokers = 0.2 # 0,2% s√£o pessoas fumantes

#Vari√°veis
flag_fumante = np.random.rand(n) < pct_smokers # < faz compara√ß√£o e retorna verdeiro ou falso
idade = np.random.normal(40, 10, n)
altura = np.random.normal(170, 10, n)
peso = np.random.normal(70, 10, n)

#Dataframe
dados = pd.DataFrame({'altura': altura, 'peso': peso, 'flag_fumante': flag_fumante})

dados['flag_fumante'] = dados['flag_fumante'].map({True: 'Fumante', False: 'N√£o Fumante'})
```

```python
dados.shape
```

1000 linhas e 3 colunas

```python
dados.head()
```

```python
sns.set(style = "ticks") #criando um estilo

sns.lmplot(x = 'altura',
           y = 'peso',
           data = dados,
           hue = 'flag_fumante',
           palette = ['tab:blue', 'tab:orange'],
           height = 7)

plt.xlabel('Altura (cm)')
plt.ylabel('Peso (kg)')
plt.title('Rela√ß√£o entre Altura e Peso de Fumantes e N√£o Fumantes')

sns.despine() #remove as bordas

plt.show()
```

A linha azul parece indicar que n√£o h√° uma rela√ß√£o entre altura e peso quando a pessoa n√£o √© fumante. Agora a linha laranja parece indicar uma rela√ß√£o positiva entre altura e peso quando a pessoa √© fumante, isso significa que quando a pessoa fuma, quanto maior a altura, maior o peso
